```python
#Step 1
#importing all libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Step 2
#Create Iris dataframe
iris = sns.load_dataset('iris')
```

```python
print(iris)
#Printing to make sure this is right
```

```python
#step 3
#Dropping the column
iris = iris.drop('species', axis= 1)
```

```python
iris.head(10)
#Checking
```

```python
#Step 4
#Creating column and summing the existing series
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head(10)
```

```python
#Step 5
#Create empty dataframe
summaryStats = pd.DataFrame()
#Populate with the mean, std, min, and max
summaryStats['Mean'] = iris.mean()
summaryStats['Std'] = iris.std()
summaryStats['Min'] = iris.min()
summaryStats['Max'] = iris.max()

summaryStats
```
